```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pickle import load
```

```python
with open("pickles/lrc1.pkl", "rb") as f:
        lr1 = load(f)

with open("pickles/lrc2.pkl", "rb") as f:
        lr2 = load(f)

with open("pickles/lrc3.pkl", "rb") as f:
        lr3 = load(f)

with open("pickles/lrc4.pkl", "rb") as f:
        lr4 = load(f)

with open("pickles/lrc5.pkl", "rb") as f:
        lr5 = load(f)
```

```python
df = pd.DataFrame.from_dict({
    "ts": lr1["ts"],
    "C1=C2=C3=C4=C5": lr1["traffic"]
})

df = pd.melt(df, id_vars=["ts"], value_vars=["C1=C2=C3=C4=C5"], var_name="DS", value_name="Traffic")

sns.lineplot(df, x="ts", y="Traffic", hue="DS")
plt.legend(loc="center")
plt.xticks(range(12, 168, 24), labels=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"])
plt.xticks(rotation=45)
plt.xlabel("")
plt.ylabel("Average number of cars")
plt.show()
```

```python
with open("simulation_arrays/ffnn1.pkl", "rb") as f:
    ffnn1 = load(f)

with open("simulation_arrays/ffnn2.pkl", "rb") as f:
    ffnn2 = load(f)

with open("simulation_arrays/ffnn3.pkl", "rb") as f:
    ffnn3 = load(f)

with open("simulation_arrays/ffnn4.pkl", "rb") as f:
    ffnn4 = load(f)

with open("simulation_arrays/ffnn5.pkl", "rb") as f:
    ffnn5 = load(f)

with open("simulation_arrays/ffnn6.pkl", "rb") as f:
    ffnn6 = load(f)
```

```python
def plot_ffnn(df):
    sns.lineplot(data=df, x="ts", y="traffic")
    plt.xticks(range(12, 168, 24), labels=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"], rotation=45)
    plt.ylabel("Median Number of Cars")
    plt.xlabel("")
    plt.show()
    
```

```python
plot_ffnn(ffnn1)
plot_ffnn(ffnn2)
plot_ffnn(ffnn3)
plot_ffnn(ffnn4)
plot_ffnn(ffnn5)
plot_ffnn(ffnn6)
```
